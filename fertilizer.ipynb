```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation,Dropout
```

```python
df = pd.read_csv('Fertilizer Prediction.csv')
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
lb_make = LabelEncoder()
df["label"] = lb_make.fit_transform(df["Fertilizer Name"])
print('done')
```

```python
df
```

```python
from sklearn.preprocessing import OneHotEncoder
df = pd.get_dummies(df, columns = ['Soil Type', 'Crop Type']) 
```

```python
df.columns,len(df.columns)
```

```python
df.head()
```

```python
X = df.drop(['label','Fertilizer Name'],axis=1)
y = df['label']
```

```python
from sklearn.model_selection import train_test_split
```

```python
y.unique()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
model = Sequential()

model.add(Dense(units=22,input_dim=22, activation='relu'))

model.add(Dense(units=19,activation='relu'))

model.add(Dense(units=16,activation='relu'))

model.add(Dense(units=13,activation='relu'))

model.add(Dense(units=10,activation='relu'))

model.add(Dense(units=7,activation='softmax'))

model.compile(loss='sparse_categorical_crossentropy',optimizer='sgd',metrics=['accuracy'])
```

```python
model.summary()
```

```python
history = model.fit(X_train, y_train, epochs=200, batch_size=64)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
predictions = model.predict_classes(X_test)
```

```python
print(classification_report(y_test,predictions))
```
